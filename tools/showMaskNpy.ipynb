```python
import numpy as np
import matplotlib.pyplot as plt
npy_path = '../dataset/gt/01_0014-000_023.npy'  # 修改为你的路径
test_pixel_mask = np.load(npy_path, allow_pickle=True)  # allow_pickle=True 可读取对象数组
```

```python
test_pixel_mask.shape
```

```python
for i in range(len(test_pixel_mask)):
    plt.figure(figsize=(8, 4))
    plt.suptitle(f'Frame {i} - Frame mask: {test_pixel_mask[i]}')

    plt.imshow(test_pixel_mask[i], cmap='gray')
    plt.axis('off')

    plt.pause(0.05)
    plt.clf()
plt.close()
```
